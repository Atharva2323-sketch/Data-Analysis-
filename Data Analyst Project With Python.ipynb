```python
import pandas as pd
import numpy as nm
import seaborn as sns
import matplotlib.pyplot as plt

df1 = pd.read_csv("Expanded_data_with_more_features.csv")
print(df1.head())
```

```python
df2 = pd.read_csv("Original_data_with_more_rows.csv")
print(df2.head(5))
```

```python
df = df1.merge(df2)
print(df.head())
```

```python
print(df.head())
```

```python
#gender Ditrubation 
```

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = df , x = "Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

# freom the above chart we analyse that:
# no of females in the chart are more then meles in the chart

```python
gb = df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean" })
print(gb)
```

```python
plt.figure(figsize = (6,5))
plt.title("relationship between parents and students score")
sns.heatmap(gb, annot =True)
print(gb)
```

# from the above chart we have conclued that the education of parent have good impact on there scores

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean" })
print(gb1) 
```

```python
plt.figure(figsize = (6,5))
plt.title("relationship between parents and students score")
sns.heatmap(gb1, annot =True)
print(gb1)
```

#from the above chart we can conclude that there is no/niglijible impact on the
# students score due to there parents marital status
